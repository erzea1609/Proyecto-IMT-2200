```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
```

```python
df =pd.read_csv("Datos_limpio.csv")#cargamos los datos limpios
df.head()
```

En este notebook se responde la pregunta, si algun jugador tuvo una lesion previa, implica que es mas probable que se lesione nuevamente?

Usaremos nuevamente regresion logistica y regresion lasso respectivamente para responder esta pregunta

```python
X = df[["Previous_Injury_Count"]]
y = df["Injury_Next_Season"]
X = pd.get_dummies(X, drop_first=True)
#Escalamos datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#Hcemos test al 30%
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
log_reg = LogisticRegression(max_iter=500)
log_reg.fit(X_train, y_train)

y_pred1 = log_reg.predict(X_test)
y_prob1 = log_reg.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred1))
print(roc_auc_score(y_test, y_prob1))
```

```python
coef_dict = pd.DataFrame({
    "caracteristicas": X.columns,
    "coef": log_reg.coef_[0], 
    "odds_ratio": np.exp(log_reg.coef_[0]) #usamos exponencial para q sea mas claro el odd_ratio y "contrarestar" el log_Reg ya que el log odds es una
                                            #escala menos intuitiva 
}).sort_values("odds_ratio", ascending=False)

print("\n=== IMPORTANCIA DE VARIABLES ===")
print(coef_dict[coef_dict["caracteristicas"].str.contains("Injury", case=False)])
```

```python
Los deportistas que sufrieron una lesion previa son 2.4 veces mas propensos a lesionarse nuevamente si mantenemos constantes 
el resto de variables
```

```python
from sklearn.metrics import confusion_matrix
matriz_log = confusion_matrix(y_test, y_pred1)
plt.figure(figsize=(5,4))
sns.heatmap(matriz_log, annot=True, fmt='d')
plt.title("Matriz de Confusión - Regresión Logística")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

```python
Con esta matriz de confusion podemos ver q el modelo si bien pudo predecir bastantes verdaderos positivos = 71 y verdaderos negativos =91
hubo aun asi una gran parte de falsos negativos y positivos, esto se puede dar debido a la gran cantidad de factores que influyen en una lesion
usaremos ahora una regresion lasso para ver si al comparar la cantidad de lesiones previas con otras variables del df.
```

```python
lasso = LogisticRegression(penalty="l1", solver="liblinear", max_iter=500)
lasso.fit(X_train, y_train) #usamos lasso con la penalizacion L1

y_pred2 = lasso.predict(X_test)
y_prob2 = lasso.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred2))
print(roc_auc_score(y_test, y_prob2))

lasso_coef = pd.DataFrame({"caracteristica": X.columns,"coef": lasso.coef_[0],
                           "odds_ratio": np.exp(lasso.coef_[0])}).sort_values("odds_ratio", ascending=False)

print("\n   COEFICIENTES LASSO    ")
print(lasso_coef[lasso_coef["caracteristica"].str.contains("Injury", case=False)])
```

```python
matriz_lasso = confusion_matrix(y_test, y_pred2)

plt.figure(figsize=(5,4))
sns.heatmap(matriz_lasso, annot=True, fmt='d')
plt.title("Matriz de Confusión - LASSO ")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

```python
Tal como se ve en esta prediccion los valores siguen siendo casi iguales aun limitandolos mediante el uso de regresion lasso 
asi que seguimos con la misma conclusion anterior
```

```python
plt.figure(figsize=(6,4)) #Ahora mostarremos en un grafico de barra la probabilidad de lesiones futuras segun la cantidad de lesiones previas
                        #del jugador
sns.barplot(x=df["Previous_Injury_Count"], y=y, estimator=np.mean)
plt.title("Probabilidad de Lesión Futura según Lesión Previa")
plt.xlabel("Lesión Previa (0=No, 1=Sí)")
plt.ylabel("Probabilidad")
plt.show()
```

```python
Conclusion: A pesar de en nuestros modelos tener una cantidad no menor de falsos negativos y positivos, aun asi la diferencia entre
los q si tuvieron una lesion anterior y los que no y su probabilidad de lesionarse nuevamente es muy grande, asi que podemos decir que si 
influye tener lesiones previas.
```
