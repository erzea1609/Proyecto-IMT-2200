```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
```

```python
df =pd.read_csv("Datos_limpio.csv")#cargamos los datos limpios
df.head()
```

Ahora intentaremos responder a la pregunta: Como afectan al futbolista sus niveles de estres, en una posible lesion futura:

```python
Dado que es una pregunta que se basa en una variable objetivo binaria, que es saber si esta lesionado o no usaremos regresion logistica
```

```python
X = df[["Stress_Level_Score"]]
y = df["Injury_Next_Season"]
X = pd.get_dummies(X, drop_first=True)
#Escalamos datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#Hcemos test al 30%
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
log_reg = LogisticRegression(max_iter=500)
log_reg.fit(X_train, y_train)

y_pred1 = log_reg.predict(X_test)
y_prob1 = log_reg.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred1))
print(roc_auc_score(y_test, y_prob1))
```

```python
coef_dict = pd.DataFrame({
    "caracteristicas": X.columns,
    "coef": log_reg.coef_[0], 
    "odds_ratio": np.exp(log_reg.coef_[0]) #usamos exponencial para q sea mas claro el odd_ratio y "contrarestar" el log_Reg ya que el log odds es una
                                            #escala menos intuitiva 
}).sort_values("odds_ratio", ascending=False)

print("\n=== IMPORTANCIA DE VARIABLES ===")
print(coef_dict[coef_dict["caracteristicas"].str.contains("Stress", case=False)])
```

```python
Dado que el coef es de 1.46 nos indica que si hay un riesgo de lesion mayor mientras mayor estres tenga el jugador, siendo mas especificos con el 
odds_ratio por cada punto de stres que aumenta en el jugador , su riesgo de lesion aumenta 4.34 veces
```

```python
from sklearn.metrics import confusion_matrix
matriz_log = confusion_matrix(y_test, y_pred1)
plt.figure(figsize=(5,4))
sns.heatmap(matriz_log, annot=True, fmt='d')
plt.title("Matriz de Confusión - Regresión Logística")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

```python
Ahora aplicaremos una regresion lasso para confirmar lo que antes habiamos predicho , ya que la regresion lasso nos dira si el estres sigue siendo
relevante aun cuando lo comparamos con otras variables de nuestro dataset
```

```python
lasso = LogisticRegression(penalty="l1", solver="liblinear", max_iter=500)
lasso.fit(X_train, y_train) #usamos lasso con la penalizacion L1

y_pred2 = lasso.predict(X_test)
y_prob2 = lasso.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred2))
print(roc_auc_score(y_test, y_prob2))

lasso_coef = pd.DataFrame({"caracteristica": X.columns,"coef": lasso.coef_[0],
                           "odds_ratio": np.exp(lasso.coef_[0])}).sort_values("odds_ratio", ascending=False)

print("\n   COEFICIENTES LASSO    ")
print(lasso_coef[lasso_coef["caracteristica"].str.contains("Stress", case=False)])
```

```python
matriz_lasso = confusion_matrix(y_test, y_pred2)

plt.figure(figsize=(5,4))
sns.heatmap(matriz_lasso, annot=True, fmt='d')
plt.title("Matriz de Confusión - LASSO ")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()
```

```python
Tal como podemos ver lasso confirma que el estres si tiene una relevancia muy grande en una posible lesion futura, tal como podemos 
observar en el sig grafico que compara ambas regresiones
```

```python
stress_range = np.linspace(df["Stress_Level_Score"].min(), df["Stress_Level_Score"].max(), 100)

# Crear dataset con solo el estres
X_temp = pd.DataFrame({"Stress_Level_Score": stress_range})

for col in X.columns:
    if col not in X_temp:
        X_temp[col] = 0

X_temp_scaled = scaler.transform(X_temp)

# Probabilidad predicha por ambos modelos
prob_log = log_reg.predict_proba(X_temp_scaled)[:, 1]
prob_lasso = lasso.predict_proba(X_temp_scaled)[:, 1]

plt.figure(figsize=(10,5))
plt.plot(stress_range, prob_log, label="Logística", linewidth=2)
plt.plot(stress_range, prob_lasso, label="LASSO", linewidth=2)
plt.xlabel("Nivel de Estrés")
plt.ylabel("Probabilidad de Lesión")
plt.title("Probabilidad de Lesión según Nivel de Estrés")
plt.legend()
plt.grid(True)
plt.show()
```

```python
Finalmente para responder esta pregunta, si podemos decir que el stres es un factor muy relevante en las posibles lesiones
futuras que pueda padecer un futbolista
```
