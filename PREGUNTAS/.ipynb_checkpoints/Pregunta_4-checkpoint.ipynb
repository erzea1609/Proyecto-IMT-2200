```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix ,classification_report 
```

# En este notebook ,gracias a los datos limpios que logramos generar,tratamos de contestar la pregunta :

#### -¿Los jugadores con mejores resultados en test físicos son más propensos a lesionarse?

```python
datos =pd.read_csv("Datos_limpio.csv")#cargamos los datos limpios
datos.head()
```

```python
datos.info() ##vemos las columnas que tenemos en nuestros DataFreame
```

```python
###---pregunta 1: ----¿Los jugadores con mejores resultados en test físicos son más propensos a lesionarse--------
tests= datos.iloc[:, 7:13] ### separamos todos las columnas que correpondan a tests
lesiones = datos["Injury_Next_Season"] ## consideramos solo las lesiones futuras
```

```python
tests.head()
```

```python
lesiones.head()
```

#### Como los datos de tiempo de reaccion y el test de velocidad se considera mejor puntaje ,el que demora menos tiempo,entonces se hara la tranformacion en ambas columna,dejando los valores en negativo, esto con el proposito que si una persona tiene mayor puntaje en estas pruebas,quiere decir que su rendimiento no es tan bueno,por lo que se "castigaria" con este puntaje.Esto se realizara,con el proposito de que cuando calculemos el promedio de todas las pruebas,realmente sea con puntajes coherentes.

```python
tests[["Reaction_Time_ms", "Sprint_Speed_10m_s"]] = tests[["Reaction_Time_ms", "Sprint_Speed_10m_s"]].apply(func=lambda x : -x)
```

```python
tests.head()#comprobamos que el cambio se realizo bien
```

```python
tests["Puntaje_promedio"]= tests.mean(axis=1)##Almaceno el promedio de los test como nueva columna
```

```python
tests.head()
```

```python
##--- ORDENAMOS POR EL PUNTAJE OBTENIDO ----- 
tests = tests.sort_values(by =["Puntaje_promedio"], ascending=False)
tests.head()
```

```python
sns.histplot(x=tests["Puntaje_promedio"], kde=True)
```

```python
datos_unidos = pd.concat([tests, lesiones], axis=1)
datos_unidos
```

```python
mean_agrupo = datos_unidos.groupby("Injury_Next_Season")["Puntaje_promedio"].mean()#se intenta aproximar el promedio de los lesionados de la sgte temporada.
mean_agrupo
```

```python
sns.barplot(mean_agrupo)
plt.xlabel(mean_agrupo.name)
plt.ylabel("Promedios de puntuaciones")
```

### Debido a la informacion que obtuvimos de acuerdoo al ultimo grafico, podemos notar que exite una amplia diferencia y relacion entre aquellos que se lesionaron,podemos notar que los jugadores que se lesionan suelen tener un promedio menor en las pruebas fisicas ,lo que puede ser un importante factor que determina la probabilidad de lesion de un jugador.En las sgtes celdas veremos si existe alguna prueba que puede tener mas incidencia en las lesiones.

```python
informacion_means = []
for i in datos_unidos.columns : 
    if i != "Puntaje_promedio" and (i != "Injury_Next_Season") : 
        print(f"PROMEDIOS DE {i.upper()} ")
        mean_agrupados = (datos_unidos.groupby("Injury_Next_Season")[i].mean())
        print(mean_agrupados)
        informacion_means.append(mean_agrupados)
```

### El patron que podemos identificar es que las personas que no se lesionan tienen un promedio en test mas altos que los que si se lesiona ,por lo que ,la respuesta a la pregunta "¿Los jugadores con mejores resultados en test físicos son más propensos a lesionarse?" :NO, las personas con mejor puntuaciones en los test no son mas propensos a lesionarse,al contrario ,son menos propenso ,esto ya que su cuerpo esta acostumbrado al alto rendimiento y a poder exigirse mas de la cuenta,por lo que al jugar los partidos su cuerpo ya estaria acostumbrado al estimulo que este demanda.

### Ahora evaluaremos si es posible crear un machine learning que de acuerdo a las puntuaciones de los jugadores pueda predecir si esta se lesionara o no...

### Para ello evaluaremos diferentes modelos y escogeremos el que mejor resultado nos entregue 

```python
from sklearn.model_selection import train_test_split ,GridSearchCV ,KFold 
from sklearn.linear_model import LogisticRegression 
X_train ,X_test ,y_train , y_test = train_test_split(tests, lesiones, test_size = 0.2, shuffle = True, random_state = 42)
```

```python
X_train 
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
X_train_esc = scaler.fit_transform(X_train)
X_test_esc = scaler.fit_transform(X_test)
```

### Escalamos lo datos para que el modelo no genere un sesgo por la diferencia de valores,por ejemplo,la columna de tiempo de reaccion son valores muy altos,por lo que el modelo puede predecir erroneamente

```python
log_reg = LogisticRegression() 
params_rl = {"C": [0.1, 0.001, 1, 10]}
kf_lr = KFold(n_splits=6, shuffle=True , random_state=42)

prueba_lr= GridSearchCV(log_reg, params_rl, cv = kf_lr )
prueba_lr.fit(X_train_esc , y_train)
print(prueba_lr.best_params_, prueba_lr.best_score_)
```

### En la celda anterior se hizo uso de los metodos de GridSearchCV y KFold, el primero lo que hace es que recibe un modelo para entrenar ,un conjunto de parametros y un kfold.Esta funcion lo que hace es generar las posibles combinaciones posibles con los paramtros,a su vez el KFold genera la validacion para que se elija el modelo que tiene el mejor promedio en la validacion cruzada y ademas nos devuelve los parametros que hace que el modelo tenga ese comportamiento. 

### Este procedimieto se generara por cada modelo para encontrar que hiperparametros hace que el modelo funcione mejor

# Probaremos los modelos con datos escalados y no escalados

```python
from sklearn.ensemble import RandomForestClassifier
# ...random forest no escalados----
random_forest =RandomForestClassifier() 
k_f_r = KFold(n_splits=6, shuffle=True , random_state=42)
params = {"n_estimators":[100,300 ,500], 
          "max_depth" :[10,15,20]}
grilla_random = GridSearchCV(random_forest, params, cv = k_f_r )
grilla_random.fit(X_train , y_train)
print(grilla_random.best_params_, grilla_random.best_score_)
```

```python
# random forest escalados
random_forest =RandomForestClassifier() 
k_f_r = KFold(n_splits=6, shuffle=True , random_state=42)
params = {"n_estimators":[100,300 ,500], 
          "max_depth" :[10,15,20]}
grilla_random = GridSearchCV(random_forest, params, cv = k_f_r )
grilla_random.fit(X_train_esc , y_train)
print(grilla_random.best_params_, grilla_random.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
#---knn  escaldos
knn = KNeighborsClassifier()
params={"n_neighbors":np.arange(1,30)}
kfold_knn =KFold(n_splits=6, shuffle=True, random_state=42)
grid_knn = GridSearchCV(knn ,params, cv=kfold_knn)
grid_knn.fit(X_train_esc, y_train)
print(grid_knn.best_params_, grid_knn.best_score_)
```

```python
#--knn sin escalar---
knn = KNeighborsClassifier()
params={"n_neighbors":np.arange(1,30)}
kfold_knn =KFold(n_splits=6, shuffle=True, random_state=42)
grid_knn = GridSearchCV(knn ,params, cv=kfold_knn)
grid_knn.fit(X_train, y_train)
print(grid_knn.best_params_, grid_knn.best_score_)
```

## Usamos los datos ecalados para knn .ya que este modelo se basa en distancias,por lo que valores con grandes distancias influyenen  las desiciones de este...

```python
tree =DecisionTreeClassifier()
kfold_tree = KFold(n_splits=7, shuffle=True, random_state=42)
params_tree ={"max_depth":np.arange(1,10),
    "min_samples_split": np.arange(2,5),
    "min_samples_leaf" : np.arange(1,5)}

grilla_trre = GridSearchCV(tree, params_tree, cv = kfold_tree)
grilla_trre.fit(X_train, y_train)
print(grilla_trre.best_params_, grilla_trre.best_score_)
```

```python
#arbol escalados ---
tree =DecisionTreeClassifier()
kfold_tree = KFold(n_splits=7, shuffle=True, random_state=42)
params_tree ={"max_depth":np.arange(1,10),
    "min_samples_split": np.arange(2,5),
    "min_samples_leaf" : np.arange(1,5)}

grilla_trre = GridSearchCV(tree, params_tree, cv = kfold_tree)
grilla_trre.fit(X_train_esc, y_train)
print(grilla_trre.best_params_, grilla_trre.best_score_)
```

### Como podemos notar,el modelo que tiene mas puntaje es el de Regresion Logistica sin escalar  por lo que entrenaremos el modelo y  analizaremos sus metricas ...

```python
logistic_regression_final = LogisticRegression(C=1)
logistic_regression_final.fit(X_train_esc, y_train)
print("SCORE TRAIN",logistic_regression_final.score(X_train_esc, y_train))
print("SCORE TEST", logistic_regression_final.score(X_test_esc, y_test))
```

```python
y_pred = logistic_regression_final.predict(X_test_esc)
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

### Como podemos notar el modelo tiene la misma probabilidad de acierto que si tiraramos una moneda al aire y escogiramos segun como caiga, esto se debe a que no detecta patrones utiles en los datos .Como podemos observar de las metricas ,el modelo "acierta" todas las que son verdadero , esto ya que adivina si la categoria es True o False ,y como  estan en la misma cantidad , detecta patrones importantes con los que pueda mejorar sus aciertos.
### Por lo que en aspectos tecnicos, no fue posible generar un modelo que fuera capaz de detectar segun datos si una persona se va a lesionar,pero si fue posible estimar valores en los cuales las personas son mas propensas a lesionarse.
